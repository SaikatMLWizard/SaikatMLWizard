```python
#Importing the Lake_20201001.txt and Inserting the data into the LakeData Table using the sqlite3 Library according to
# the task given
import sqlite3

import csv
#from csv import unicodecsv


con = sqlite3.connect('mycorpus.db')
cur = con.cursor()
cur.execute("DROP TABLE LakeData")
cur.execute("CREATE TABLE LakeData(Application Text, Date Datetime, Amount	Numnber,Load_Date Date)")

with open('Lake_20201001.txt', 'rt') as input_file:
    reader = csv.reader(input_file, delimiter="\t")
    data = [row for row in reader]

cur.executemany("INSERT INTO LakeData(Application , Date , Amount,Load_Date ) VALUES (?, ?, ?, ?);", data)

con.commit()
```

```python
#Checking the Data being Inserted into the Table is accordance with the original data

cur.execute("SELECT * FROM LakeData"); 
rows = cur.fetchall()
#con.close()
for row in rows:
   print(row)
```

```python
#Random Filtering of the Data based upon the Date

cur.execute("SELECT * FROM LakeData where Date> '2010-01-01' AND Date < '2010-12-31'")
rows = cur.fetchall()
#con.close()
for row1 in rows:
   print(row1)

con.commit()
```
